```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def displayImage(image, title): 
    plt.imshow(image)
    plt.title(title)
    plt.show()
```

```python
image = cv2.imread('./data/fish_test.png')
displayImage(image, "original fish")
```

```python
p = np.ones(image.shape, dtype="uint8") * 100
increased = cv2.add(image, p)
displayImage(increased, "increased fish")
```

```python
m = np.ones(image.shape, dtype="uint8") * 60
decreased = cv2.subtract(image, m)
displayImage(decreased, "decreased fish")
```
